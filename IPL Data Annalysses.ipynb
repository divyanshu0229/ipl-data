<h4> 1. Matches won by maximum margin of runs.</h4>

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("matches.csv")

max_margin = df['win_by_runs'].max()

max_match = df[df['win_by_runs'] == max_margin].iloc[0]

winner = max_match['winner']
team1 = max_match['team1']
team2 = max_match['team2']

opponent = team2 if winner == team1 else team1

label = f"{winner} vs {opponent}"

plt.figure()
bars = plt.bar(label, max_margin)

plt.xlabel("Match",fontweight='bold',fontsize=25)
plt.ylabel("Runs",fontweight='bold',fontsize=25)
plt.title("Maximum Margin of Victory (Runs)",fontweight='bold',fontsize=25)

plt.text(label, max_margin, str(max_margin), 
         ha='center', va='bottom', fontsize=12)

plt.xticks(rotation=0,fontweight='bold',fontsize=15)
plt.yticks(rotation=0,fontweight='bold',fontsize=15)
plt.tight_layout()

plt.show()
```

<h4> Insights </h4>

- The highest victory margin is 146 runs.
- It indicates a highly one-sided match
- The winning team (Mumbai Indians) completely dominated the opponent (Delhi Daredevils).
- This is an exceptional performance compared to other matches.

- 48% of the matches are won after losing the toss. It means you can win the match after losing the toss.

<h4>15.Best venue for defending and chasing a total.</h4>

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("matches.csv")

defend = df[df['win_by_runs'] > 0]
defend_counts = defend['venue'].value_counts()

chase = df[df['win_by_wickets'] > 0]
chase_counts = chase['venue'].value_counts()

venue_comparison = pd.DataFrame({
    'Defending Wins': defend_counts,
    'Chasing Wins': chase_counts
}).fillna(0)

venue_comparison['Total'] = venue_comparison.sum(axis=1)
venue_comparison = venue_comparison.sort_values('Total', ascending=False).head(10)

plt.figure(figsize=(30,18))

x = range(len(venue_comparison))
plt.bar(x, venue_comparison['Defending Wins'], width=0.4, label='Defending Wins')
plt.bar([i + 0.4 for i in x], venue_comparison['Chasing Wins'], width=0.4, label='Chasing Wins')

plt.xticks([i + 0.2 for i in x], venue_comparison.index, rotation=45)
plt.xlabel("Venue",fontsize = 40,fontweight='bold')
plt.ylabel("Number of Matches",fontsize = 35,fontweight='bold')
plt.title("Best Venues for Defending vs Chasing (Top 10)",fontsize = 40,fontweight='bold')
plt.legend()

plt.xticks(rotation=90,fontsize = 25)
plt.yticks(rotation=90,fontsize = 25)
plt.tight_layout()
plt.show()
```

<h4>Insights</h4>

- Eden Gardens is easy for chasing teams, chasing wins (45) are much higher than defending wins (32), indicating teams prefer batting second here.
- Feroz Shah Kotla shows equal results for both defending and chasing wins are equal (33 each), making it a neutral venue.
- Only MA chidambaram stadium have higher defending wins than chasing wins means in these top 10 venues only this stadium favors defending wins.
- Sawai mansingh stadium have large difference in defending and chasing wins shows this stadium highly favors batting second.

<h4>16.Best defending and chasing teams.</h4>

```python
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_csv("matches.csv")

defend = df[df['win_by_runs'] > 0]
defend_counts = defend['winner'].value_counts()

chase = df[df['win_by_wickets'] > 0]
chase_counts = chase['winner'].value_counts()

winner_comparison = pd.DataFrame({
    'Defending Wins': defend_counts,
    'Chasing Wins': chase_counts
}).fillna(0)

winner_comparison['Total'] = winner_comparison.sum(axis=1)

winner_comparison = winner_comparison.sort_values('Total', ascending=False).head(10)

plt.figure(figsize=(14,8))

x = range(len(winner_comparison))

plt.bar(x, winner_comparison['Defending Wins'], width=0.4, label='Defending Wins')
plt.bar([i + 0.4 for i in x], winner_comparison['Chasing Wins'], width=0.4, label='Chasing Wins')

plt.xticks([i + 0.2 for i in x], winner_comparison.index, rotation=45)
plt.xlabel("Teams",fontsize = 25,fontweight='bold')
plt.ylabel("Number of Matches",fontsize = 25,fontweight='bold')
plt.title("Best Teams for Defending vs Chasing (Top 10)",fontsize = 28,fontweight='bold')
plt.legend()

plt.xticks(rotation=90,fontsize = 16)
plt.yticks(rotation=0,fontsize = 16)
plt.tight_layout()
plt.show()
```

<h4>Insights</h4>

- Only some teams have higher defending wins than chasing wins like MI, CSK, SRH and DC. This shows most teams performs good when they are chasing.
- Data of teams like MI, CSK, KXIP and SRH have performed good in both chasing and defending it shows their team is prepared for bat or field first.
- The graph shows Gujrat Lions performance is very poor when they are defending the runs.
- Teams like KKR, RR and DD have very good performance when they are chasing as compared to defending the totals.
